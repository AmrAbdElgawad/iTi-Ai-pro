```python
#Write a NumPy code line(s) to add elements in a matrix. If an element in the matrix is 0,
#we will not add the element below this element (in red)
#M= 1 1 0  0 3 0 1 0 4    2     3    4 
arr=np.array([[1,1,0,0],
             [3,0,1,0],
             [4,2,3,4]])
sumi =0
x=[]

# to get indexes of values that are below 0 in same column
for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if arr[i][j]==0 and i<arr.shape[0]-1:
            x.append([i+1,j])
# replace the values in these indexes by 0 si its not added in sum func
for c,v in x:
    arr[c,v]=0

print(arr)
# using np.sum to add the whole elements in array after removing elements that is below 0
print(np.sum(arr))
```

```python
# Write a NumPy code line(s) to extract all numbers which are less and greater than a specified integer in an input array
aray=np.array([2,5,4,44,5,66,9,5,4,55])

n=10

print(" elements less than ",n ," : ",end='')
for elmnt in aray:
    if elmnt<n:
        print(elmnt," , ",end='')
        
print("\nelements greater than ",n ," : ",end='')
for elmnt in aray:
    if elmnt>n:
        print(elmnt," , ",end='')

```
